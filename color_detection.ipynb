```python
import cv2 as cv
import matplotlib.pyplot as plt
from PIL import Image
img = cv.imread("/content/drive/MyDrive/1102813.jpg")
```

```python
blur = cv.blur(img,(5,5))
blur0=cv.medianBlur(blur,5)
blur1= cv.GaussianBlur(blur0,(5,5),0)
blur2= cv.bilateralFilter(blur1,9,75,75)
```

```python
hsv = cv.cvtColor(blur2, cv.COLOR_BGR2HSV)
```

```python
#for blue color
import numpy as np
low_blue = np.array([70, 50, 50])
high_blue = np.array([110, 255, 255])
mask = cv.inRange(hsv, low_blue, high_blue)
```

```python
res = cv.bitwise_and(img,img, mask= mask)
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(res)
```
